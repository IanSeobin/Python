# 의사결정나무

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.tree import DecisionTreeClassifier # 의사결정나무

# 의사결정나무 모델 생성
decision_tree = DecisionTreeClassifier(random_state=32)

# 의사결정 모델 학습
decision_tree.fit(X_train, y_train)

# test 데이터로 예측하고 정답 확인
y_pred = decision_tree.predict(X_test)

#예측, 정답
y_pred, y_test
```

```python
# 혼동행렬(confusion matrix) 그려보기
from sklearn.metrics import confusion_matrix

# 예측, 실제 데이터 혼동행렬 가져오기
cm = confusion_matrix(y_test, y_pred)
```

```python
# 혼동 행렬 시각화
plt.figure(figsize=(8,6))

# x축: 예측(predict), y축: 실제(test)
sns.heatmap(cm, annot=True, fmt='d', cmap='Reds', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel("Predicted Labels")
plt.ylabel("True Lables")
plt.title("Confusion Matrix")

plt.show()
```

```python
from sklearn.metrics import accuracy_score

# 의사결정나무 모델 정확도(Accuracy) 평가 (의사결정 나무 모델을 학습시켜야 함)
accuracy = accuracy_score(y_test, y_pred) # 정확도 평가
accuracy
```

```python
from sklearn.metrics import classification_report

# accuracy(정확도), precision(정밀도), recall(재현율) 
# f1-score(f1스코어: precision과 recall의 가중 조화평균), support(각 라벨의 샘플개수) 확인
print(classification_report(y_test, y_pred))
```

# 랜덤 포레스트

```python
from sklearn.ensemble import RandomForestClassifier # 랜덤포레스트

# 랜덤포레스트 모델 생성
random_forest = RandomForestClassifier(random_state=32)

# 랜덤포레스트 모델 학습
random_forest.fit(X_train, y_train)

# test 데이터로 예측하고 정답 확인
y_pred = random_forest.predict(X_test)

# 예측, 정답
y_pred, y_test
```

```python
# 랜덤 포레스트 모델 정확도(Accuracy) 평가 (랜덤 포레스트 모델을 학습시켜야 함)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
